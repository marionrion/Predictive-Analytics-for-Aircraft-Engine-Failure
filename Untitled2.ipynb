# Predictive Analytics for Aircraft Engine Failure

## Description

The goal of this project is to develop a predictive analytics model to forecast aircraft engine failures. By analyzing various sensor readings and operational settings, the model will help in predicting the time-to-failure (TTF) of engines, thereby improving maintenance scheduling and reducing unexpected failures.

## Experimental Scenario

The dataset used in this study simulates sensor data collected from aircraft engines. This data encompasses different operational settings, sensor measurements, and failure indicators, enabling us to predict engine failures accurately. The approach involves preprocessing the data, selecting features, training various machine learning models, and evaluating their performance.

## The Dataset

The dataset contains multiple types of files, each representing different aspects of engine operations and failures:

1. **Sensor Data**: Raw sensor measurements taken from the engines.
2. **Operational Settings**: Parameters that control the engines operating conditions.
3. **Failure Labels**: Indicators of engine failures and their time-to-failure (TTF).

### Feature Information

The dataset includes the following columns:

- **Identifiers**:
  - `id`: Unique identifier for each engine.
  - `cycle`: Operating cycle number.

- **Operational Settings**:
  - `setting1`, `setting2`, `setting3`: Various settings influencing engine performance.

- **Sensor Measurements**:
  - `s1` to `s21`: Sensor readings capturing engine conditions.

- **Averages of Sensor Measurements**:
  - `av1` to `av21`: Average sensor readings over certain periods.

- **Standard Deviations of Sensor Measurements**:
  - `sd1` to `sd21`: Standard deviation of sensor readings indicating variability.

- **Target Variable**:
  - `ttf`: Time-to-failure of the engine.

- **Labels**:
  - `label_bnc`: Binary failure label.
  - `label_mcc`: Multiclass failure label.

## Problem Statement

The primary objective is to develop a predictive model that can accurately forecast the remaining time-to-failure (TTF) of aircraft engines based on historical sensor and operational data. This prediction will enable proactive maintenance strategies and reduce the risk of unexpected engine failures.

## Gradient Boosting

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create the Gradient Boosting model
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Train the model
gb_model.fit(x_train, y_train)

# Predict on test data
y_pred = gb_model.predict(x_test)
```

```python
# Evaluate the model
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {MAE}")
print(f"Mean Squared Error: {MSE}")
print(f"Root Mean Squared Error: {RMSE}")
print(f"R-squared: {R2}")
```
